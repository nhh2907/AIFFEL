### Dates 변환
- Dates 피처를 년, 월, 일, 시, 분 으로 나누어서 각각 컬럼에 저장

```python
# 날짜를 변환해서 년원일분으로 나눠주는 함수
def split_date(df):
    date = pd.to_datetime(df["Dates"])
    df["Dates"] = date
    df["Year"] = date.dt.year
    df["Month"] = date.dt.month
    df["Day"] = date.dt.day
    df["Hour"] = date.dt.hour
    df["Min"] = date.dt.minute
```

```python
split_date(train)
# split_date(test)
```

### 평일/주말 구분

```python
# 날짜를 받아서 주말 평일을 판단하는 함수
# 평일이면 0 주말이면 1 반환
from datetime import datetime, date

def is_weekend(d):
    if d.weekday() > 4:
        wd = "weekend"
    else:
        wd = "weekday"
    return wd
```

```python
train["is_Weeked"] = train["Dates"].apply(is_weekend)
```

### 계절 구분
- Month를 기준으로 계절 구분
- 컬럼명은 Season
- spring = [3,4,5]
- summer = [6,7,8]
- fall = [9,10,11]
- winter = [12,1,2]

```python
def season_div(x):
    if 3<=x<6:
        a = "spring"
    elif 6<=x<9:
        a = "summer"
    elif 9<=x<12:
        a = "fall"
    else:
        a = "winter"
    return a
```

```python
train["Season"] = train["Month"].apply(season_div)
```
